```python
# Keep only rows where 'book_table' is 'Yes' or 'No'
df_cleaned = df_cleaned[df_cleaned['book_table'].isin(['Yes', 'No'])]
```

```python
df_cleaned.columns
```

```python
df_cleaned=df_cleaned[df_cleaned['online_order'].isin(['Yes','No'])]
```

```python
df_cleaned['rate'].unique()
```

```python
# Remove '/5' and convert to float
df_cleaned['rate'] = df_cleaned['rate'].astype(str).str.replace('/5', '').str.strip()
```

```python
# Replace 'NEW' and '-' with NaN
df_cleaned['rate'].replace(['NEW', '-'], pd.NA, inplace=True)
```

```python
df_cleaned['rate'] = pd.to_numeric(df_cleaned['rate'], errors='coerce')
```

```python
df_cleaned['rate'] = df_cleaned['rate'].replace(['NEW', '-'], pd.NA)
```

```python
df_cleaned.reset_index(drop=True, inplace=True)
```

```python
df_cleaned.head()
```

```python
# Remove '_x000D_' from the phone number column
df_cleaned['phone'] = df_cleaned['phone'].astype(str).str.replace('_x000D_', '', regex=False)

# Optional: Strip leading/trailing spaces
df_cleaned['phone'] = df_cleaned['phone'].str.strip()
```

```python
df_cleaned.tail()
```

```python
for col in df_cleaned.columns:
    if df_cleaned[col].astype(str).str.contains(r"\[\]").any():
        df_cleaned.drop(columns=col, inplace=True)
```

```python
df_cleaned.head()
```

```python
df_cleaned['phone'] = (
    df_cleaned['phone']
    .str.replace(r'\r|\n|\t', '', regex=True)   # remove hidden characters
    .str.replace(r'\s{2,}', ' ', regex=True)    # remove extra spaces
    .str.strip()
)
```
